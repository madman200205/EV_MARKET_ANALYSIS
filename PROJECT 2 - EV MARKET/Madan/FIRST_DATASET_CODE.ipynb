```python
import pandas as pd

file_path = 'EV_India.csv'

df = pd.read_csv(file_path)

# DROPING THE UNAVAILABLE DATA
df = df.dropna(subset=['Total Electric Vehicle', 'Total Non-Electric Vehicle', 'Total'])


# Convert columns to numerical values by removing commas
df['Total Electric Vehicle'] = df['Total Electric Vehicle'].str.replace(',', '').astype(int)
df['Total Non-Electric Vehicle'] = df['Total Non-Electric Vehicle'].str.replace(',', '').astype(int)
df['Total'] = df['Total'].str.replace(',', '').astype(int)

# Calculate the percentage

df['percentage'] = df['Total Electric Vehicle'] / df['Total']
print("\n\n state name with percentage of EV Vehilces")

print(df[["State Name","percentage"]])
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'EV_India.csv'

df = pd.read_csv(file_path)

# DROPING THE UNAVAILABLE DATA
df = df.dropna(subset=['Total Electric Vehicle', 'Total Non-Electric Vehicle', 'Total'])

# Clean up column names by removing leading/trailing whitespaces
df.columns = df.columns.str.strip()

# Convert columns to numerical values by removing commas
df['Total Electric Vehicle'] = df['Total Electric Vehicle'].str.replace(',', '').astype(int)
df['Total Non-Electric Vehicle'] = df['Total Non-Electric Vehicle'].str.replace(',', '').astype(int)

# Plotting the data on separate axes
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10), sharex=True)

# Plotting Total Electric Vehicle on the first axis
ax1.bar(df['State Name'], df['Total Electric Vehicle'], label='Total Electric Vehicle', color='blue')
ax1.set_ylabel('Number of Electric Vehicles')
ax1.set_yscale('log')
ax1.legend()

# Plotting Total Non-Electric Vehicle on the second axis
ax2.bar(df['State Name'], df['Total Non-Electric Vehicle'], label='Total Non-Electric Vehicle', color='orange')
ax2.set_ylabel('Number of Non-Electric Vehicles')
ax2.set_yscale('log')
ax2.legend()

plt.xticks(rotation=90)
plt.xlabel('State Name')
plt.title('Total Electric and Non-Electric Vehicles for Each State')
plt.tight_layout()
plt.show()
```

States with Higher EV Usage:

Uttar Pradesh has the highest number of electric vehicles (3,37,180), followed by Delhi (1,56,393), Karnataka(1,20,532),Maharashtra	(1,16,646) and Bihar(83335).
The top five states with higher EV usage (in terms of the number of electric vehicles) are: Uttar Pradesh, Delhi, Karnataka, Maharashtra, and Bihar.

States with Lower EV Usage:

Arunachal Pradesh has the lowest number of electric vehicles, followed by Sikkim and Mizoram (21), Ladakh(26), and Meghalaya (49).
The bottom five states with lower EV usage are Arunachal Pradesh, Mizoram, Sikkim, Ladakh, and Meghalaya.


States with Lower EV Usage but Higher Non-EV Usage:

Mizoram has relatively lower electric vehicle usage compared to some other states but has a higher number of non-electric vehicles.
Similarly, Arunachal Pradesh and Meghalaya also have lower EV usage but a relatively higher number of non-electric vehicles.

States with Higher EV Usage and Lower Non-EV Usage:

Delhi has relatively higher electric vehicle usage compared to some states and lower non-electric vehicle usage.
Tripura, Uttar Pradesh, and Karnataka also have higher EV usage and lower non-electric vehicle usage.
States with Balanced EV and Non-EV Usage:

Some states like Assam, Gujarat, Haryana, Kerala, Odisha, Punjab, and West Bengal have a significant number of electric vehicles  and a considerable number of non-electric vehicles.
States with Missing Data:

Some states like Andra Pradesh, Lakshadweep, Madhya Pradesh, and Telangana have missing data for both electric and non-electric vehicles.
